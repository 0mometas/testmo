```python
name = input("Input Yield/Table name(.xlsx or .xlsm) : ")
yield170 = pd.read_excel(name)
yield170.head()
```

```python
print("shape = ",end='')
print(yield170.shape,sep = None,end='\n\n')
yield170.info()
```

```python
lot170 = []
yieldTCI = []
out_date = []
mcid = []
rate_adj = []

for lot in yield170['Lot No.']:
    lot170.append(lot)
for y in yield170['Yield']:
    yieldTCI.append(y)
for d in yield170['Out Date']:
    out_date.append(d)
for ID in yield170['TCI'] : 
    mcid.append(str(ID))
for adj in yield170['Rate ADJUST NG170'] : 
    rate_adj.append(adj)

#Create DataFrame as date we use
df = pd.DataFrame(list(zip(yieldTCI,out_date,mcid,rate_adj)), index = lot170 , columns = ['yield','Date','M/C ID','ADJUST NG'])
df.sort_values('Date')
df['M/C ID'] = df['M/C ID'].str[-2:]
df['ADJUST NG'] = df['ADJUST NG'].fillna(value=0)
df
```
